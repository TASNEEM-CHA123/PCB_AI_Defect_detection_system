# Industrial PCB AOI System: Training Pipeline (YOLO26-S)

This notebook trains a YOLO26-S model on the **PKU-Market-PCB** dataset for automated optical inspection of PCBs.

## Stage 1: Environment Setup

```python
!pip install -q ultralytics roboflow
import torch
from ultralytics import YOLO
import os

print(f"Setup complete. Using torch {torch.__version__} ", f"({torch.cuda.get_device_properties(0).name if torch.cuda.is_available() else 'CPU'})")
```

## Stage 2: Download Dataset
We use the Roboflow snippet to download the PKU-Market-PCB dataset.

```python
from roboflow import Roboflow
# Placeholder: Replace with your actual Roboflow API key and project link for PKU-Market-PCB dataset
rf = Roboflow(api_key="YOUR_API_KEY_HERE")
project = rf.workspace("pcb-defects").project("pku-market-pcb")
version = project.version(1)
dataset = version.download("yolov8")
```

## Stage 3: Train YOLO26-S Model
Configure hyper-parameters for optimal industrial localization.

```python
# Utilizing the latest 'YOLO26-S' architecture (placeholder for YOLO latest small model equivalent)
model = YOLO('yolov8s.pt') # Using YOLOv8s as the proxy/base for YOLO26-S in this environment

results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name='yolo26_s_pcb_defects',
    optimizer='AdamW',
    lr0=1e-3
)
```

## Stage 4: Export Weights for Local Inference
Exporting the model to ONNX for potential CPU-accelerated deployments or utilizing PyTorch weights (.pt).

```python
modelPath = "runs/detect/yolo26_s_pcb_defects/weights/best.pt"
exported_model = YOLO(modelPath)
exported_model.export(format='onnx')
print("Training and Export Complete.")
```
